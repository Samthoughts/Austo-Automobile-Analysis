### **Data Descriptions**

- **Age:** The age of the individual in years.
- **Gender:** The gender of the individual, categorized as male or female.
- **Profession:** The occupation or profession of the individual.
- **Marital_status:** The marital status of the individual, such as married &, single
- **Education:** The educational qualification of the individual Graduate and Post Graduate
- **No_of_Dependents:** The number of dependents (e.g., children, elderly parents) that the individual supports financially.
- **Personal_loan:** A binary variable indicating whether the individual has taken a personal loan "Yes" or "No"
- **House_loan:** A binary variable indicating whether the individual has taken a housing loan "Yes" or "No"
- **Partner_working:** A binary variable indicating whether the individual's partner is employed "Yes" or "No"
- **Salary:** The individual's salary or income.
**Partner_salary:** The salary or income of the individual's partner, if applicable.
- **Total_salary:** The total combined salary of the individual and their partner (if applicable).
- **Price:** The price of a product or service.
- **Make:** The type of automobile

##**Actionable Insights & Recommendations**



#### - **Actionable Insights**



1. **Diverse Customer Segments**

Austo Motors serves distinct customer groups — families, professionals, and young buyers — each with different expectations and purchase motivations. A one-size-fits-all approach limits engagement and sales potential.

2. **SUV Dominance and Loan Dependence**

SUVs continue to be the most popular choice, especially among family-oriented buyers. Many of these customers rely on personal loans, indicating that financing options strongly influence purchasing decisions.

3. **Price Sensitivity in Hatchbacks**

Hatchback buyers are highly price-conscious and value fuel efficiency, affordability, and convenience over luxury or size.

4. **Localized Inventory Needs**

Sales success varies by region and demographic. Areas with younger populations perform better with hatchbacks, while suburban and family-oriented regions show higher SUV demand.

5. **Marketing Mismatch**

Current marketing campaigns are too general and fail to address the unique motivations of each customer segment.

#### - **Business Recommendations**

1. **Refine Marketing Messaging**

**SUVs:** Focus campaigns on family, safety, comfort, and reliability. Use media platforms that reach older and family-centric audiences.

**Sedans:** Highlight elegance, technology, and value for professionals. Use digital media, career platforms, and lifestyle magazines.

**Hatchbacks:** Emphasize economy, fun, and convenience. Use social media and youth-focused marketing.

2. **Develop Strategic Partnerships**

Partner with banks and financial institutions to offer pre-approved or low-interest car loans, especially targeting SUV buyers.

Promote these financing options in campaigns to make purchasing smoother and more appealing.

3. **Optimize Sales and Inventory**

Train sales teams to quickly identify a buyer’s profile and tailor recommendations accordingly.

Adjust dealership inventory based on local demographics:

More hatchbacks in areas with younger populations.

More SUVs in family-heavy suburbs.

4. **Guide Product Development**

Expand the SUV lineup with varied price points and model sizes to capture growing demand.

Keep hatchbacks competitively priced with strong mileage and low maintenance costs to maintain their market appeal.
